```python
import pandas as pd
```

```python
vul = pd.read_csv("vulnerability_missing_summary.csv")
read = pd.read_csv("readiness_missing_summary.csv")
```

```python
final_summary = pd.merge(vul, read, how="outer")
```

```python
final_summary = final_summary.fillna(" ").set_index("Unnamed: 0")
```

```python
final_summary.index.name = None
```

```python
final_summary['Missing Count'] = final_summary.iloc[:,1:].eq("Missing").sum(axis=1)
```

```python
final_summary = final_summary.sort_values(by="Missing Count", ascending=False)
```

```python
final_summary['Missing Ratio'] = final_summary['Missing Count']/14
final_summary
```

```python
names = pd.read_csv("../raw_data/readiness/readiness.csv")
names = names.set_index("ISO3")
```

```python
final_summary['Name'] = names['Name']
```

```python
drop_countries = final_summary[final_summary['Missing Ratio'] > 0.3]['Name']
```

```python
drop_countries = drop_countries.to_dict()
```

```python
drop_countries
```

```python
pd.Series(drop_countries, name="Name").rename_axis("ISO3").to_csv("drop_countries.csv")
```
